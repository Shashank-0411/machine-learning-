###Classification probelm
This is a classification problem where we have to predict if the first pokemon won the battle or the second pokemon won the battle. So we can say that this is a binary classification problem because the outcome can be only one of the two. To represnet this binary classification we would change the winner column to 0 or 1, where 0 means that the first pokemon won and 1 means that the second pokemon won.


We loop through each row in the dataframe and check if the winner is first pokemon or not. If the winner is the first pokemon then we change the winner column to 0 or else we change it to 1 representing that the second pokemon has won the battle.

```
for i in range(len(combats)):
  if combats.Winner[i] == combats.First_pokemon[i]:
    combats.Winner[i] = 0
  else:
    combats.Winner[i] = 1
```

```
combats.head()
```

The above task can be done in more pythonic way. This method is shown below

First we equate winner to the first pokemon column. this would give a list of true and false. the indexes that have the winner as first pokemon would be true and other indexes would have the false value. Now we insert this 1d matrix to combats.Winner column. This would replace the indexes containing true with the value assigned, in this case it is 0.

next we check if the winner is equal to the second pokeom. this would again give a 1d matrix of true and false. true for the cases in which the second pokemon has won. now we give this 1d matrix to combats.Winner which would change the winner column with 1 wherever there is true.

Our combats file is almost ready. Now we have to map it to the stats of each pokemon. For this we need to convert the pokemon dataframe into a dictonary(key:value) where key would be the number (for 266 pokemon) and value will be its stats.

We consider all the columns except the Name column, because name is not numerical and cannot be compared

```
stats = pokemon.columns
stats = stats[1:]
stats
```

To convert the dataframe into dictonary we will use to_dict method. this method changes the column name as key and the column values as value thus forming a key:value pair.

So we need to transpose the dataframe. To transpose the dataframe use .T operator to transpose. Example is shown below.

```
pokemon[stats].T
```

Now we would apply to_dict method. key would be 1,2,3...800. and the values would be the mentioned stats. we want this stats to be in a list so inside the to_dict method we mention "list". So the final output would be a key:value pair where value contains a list of stats for each pokemon

```
stats_dict = pokemon[stats].T.to_dict("list")
stats_dict
```

Now we have our dictonary ready and for the next step we have to map this dictonary to the combats dataframe. we would access each pokemon's stat using its number as the key.

map method would do this for us. It takes a dictionary as an input and applies it to the calling dataframe. to here combats.First_pokemon is the calling dataframe and we apply map(stats_dict) to this sub-dataframe.

50000 entries are changes from just the pokemon number to their stats.

```
one = combats.First_pokemon.map(stats_dict)
one
```

As done in the above code shell we would do the same thing for the second pokemon column also.

```
two = combats.Second_pokemon.map(stats_dict)
two
```

ONE is for the first pokemon and TWO is for the second pokemon.
Now we have to compare the stats of ONE and TWO, in simple term we can say that we have to take a difference. For that we have to subtract "hit point" of senond pokemon from "hit point" of first pokemon. We have to subtract "defence" of second pokemon from "defence" of first pokemon and so on for each stats value. This is known as element-wise subtraction.

Element-wise mathematical operations can be done using numpy library. So just before applying the substraction operation we have to convert the stats for each of the pokemon into a numpy array.

Here we loop starting from 0 to 49999 and compare(subtract) TWO from ONE and save it in a list named as comparision. We define an empty list with [] and we add new data into the list using append method. So finally we have a list that contains numpy array of comparision between first and second pokemon.

```
comparision = []
for i in range(len(one)):
  comparision.append(np.array(one[i]) - np.array(two[i]))
```

```
comparision
```

We have the comparision. Now we would add the winner column to the comparision list. To add this Winner column we are coverting the comparision list into a pandas dataframe. Each comparision contains 6 entries(hit points, attack, defence, sp. atk, sp. def, speed and legnedry). So while creating the dataframe we will mention what are the column names for our understanding.

```
almost_data = pd.DataFrame(comparision, columns=stats)
almost_data.head()
```

Now we would use concat method to include the winner column in the dataframe. we want this data to be added as a column, so mention that the axis for this concatenation is 1(1 is for column and 0 is for rows).

Finally our dataframe is ready for splitting purpose.

```
data = pd.concat([almost_data, combats.Winner], axis=1)
data
```

Our dataframe is finally ready and has the comparision between the first pokemon and the second pokemon. Here -ve value means that the second pokemon was more powerful in the mention column and its magnitude defines how stronger he is. +ve value means that the first pokemon was stronger and the magnitude defines how stronger he is as compared to the secnod pokemon. Our last column shows which pokemon has won. 0 means that the first pokemon won and 1 means that the second pokemon won.
